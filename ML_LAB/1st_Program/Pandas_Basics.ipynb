```python
import pandas as pd

# -----------------------------
# PART 1: Creating a Series
# -----------------------------

r = pd.Series([10, 20, 30])
print("Series:\n", r)

s = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
print("Series:\n", s)

# -----------------------------
# PART 2: Creating CSV File
# -----------------------------
data = {
    'Name': ['A', 'B', 'C', 'D', 'E','F'],
    'Marks': [80, 90, 75, 85, 60,100]
}

df_create = pd.DataFrame(data)

# Save to CSV
df_create.to_csv("marks.csv", index=False)

# -----------------------------
# PART 3: Reading CSV File
# -----------------------------
df = pd.read_csv("marks.csv")

print("\nDataFrame:\n", df)

print("\nHead:\n", df.head())

print("\nTail:\n", df.tail())

print("\nFilter Marks > 80:\n", df[df['Marks'] > 80])

# -----------------------------
# PART 4: AND Operator
# -----------------------------
print("\nFilter Marks > 80 AND Name == 'D':\n",
      df[(df['Marks'] > 80) & (df['Name'] == 'D')])
```
