```python
import numpy as np

arr = np.array([1,2,3,4])
print("Array:", arr)

matrix = np.array([[1,2],[3,4]])
print("Matrix:\n", matrix)

print("Zeros:\n", np.zeros((2,3)))

print("Arange:", np.arange(0,10,2))

a = np.array([1,2,3])
b = np.array([4,5,6])
print("Addition:", a+b)

print("Mean:", arr.mean())
print("Standard Deviation:", arr.std())
```
