```python
import matplotlib.pyplot as plt

# -----------------------------
# 1️⃣ Bar Chart
# -----------------------------
students = ['Raju', 'Bob', 'Charlie', 'Diana', 'Eva']
marks = [80, 90, 75, 85, 60]

plt.bar(students, marks, color='skyblue')
plt.title("Student Marks")
plt.xlabel("Student")
plt.ylabel("Marks")
plt.show()
```

```python
# -----------------------------
# 2️⃣ Pie Chart
# -----------------------------
languages = ['Python', 'Java', 'C++', 'JavaScript', 'C#']
popularity = [40, 30, 10, 15, 5]

plt.pie(popularity, labels=languages, autopct='%1.1f%%', startangle=90)
plt.title("Popular Programming Languages")
plt.show()
```

```python
# -----------------------------
# 3️⃣ Histogram
# -----------------------------
data = [1,2,2,3,3,3,4,4,5]

plt.hist(data, bins=5, color='lightgreen', edgecolor='black')
plt.title("Data Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
